```python
import os
from pyspark.sql import SparkSession

# 1. Get the Master URL from the Docker Environment (or default to local if missing)
# In our docker-compose, we set SPARK_MASTER="spark://spark-master:7077"
master_url = os.environ.get('SPARK_MASTER', 'local[*]')

print(f"Connecting to Cluster at: {master_url}")

# 2. Create the Session
# NOTE: We do NOT need to pass .config("spark.jars.packages", ...) here.
# The PYSPARK_SUBMIT_ARGS env var handles that automatically!
spark = SparkSession.builder \
    .appName("Jupyter_Interactive_Session") \
    .master(master_url) \
    .getOrCreate()

# 3. Verify
print(f"Spark Version: {spark.version}")
print(f"Web UI: http://localhost:4040") # This maps to your host machine port
```
